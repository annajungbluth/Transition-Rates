# Transfer Rate Comparison - Dependence on G and Î»

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

### Transfer Rate

```python
pi = 3.14159265359
h = 6.62607004 * math.pow(10,-34) # [m2 kg / s]
hbar = 1.0545718 * math.pow(10,-34) # [m2 kg / s]
kb = 1.38064852 * math.pow(10,-23) # [m2 kg / (s2 K)]
T = 300 # [K]
```

```python
def trans_rate_G(G_kT, l_kT):
    # G_kT = list of values
    # l_kT = number 
    
    trans_rate_list = []
    
    for g_kT in G_kT:
        trans_rate = 1/(math.sqrt(l_kT)) *math.exp((-(g_kT + l_kT)**2) / (4 * l_kT))
        trans_rate_list.append(trans_rate)
        
    return trans_rate_list
```

```python
def trans_rate_l(g_kT, L_kT):
    # G_kT = number
    # L_kT = list of values 
    
    trans_rate_list = []
    
    for l_kT in L_kT:
        trans_rate = 1/(math.sqrt(l_kT)) *math.exp((-(g_kT + l_kT)**2) / (4 * l_kT))
        trans_rate_list.append(trans_rate)
        
    return trans_rate_list
```

### Transfer Rate as a function of G0

```python
GkT_list = np.arange(-20,20,0.2)
l_list = [0.1, 0.5, 2, 6, 10]
colors = ['#0015a1', '#ba0087', '#ff0057', '#ff851e', '#ffdd00']
```

```python
plt.figure(figsize=(6,4), dpi=300)

for x in range(len(l_list)):
    plt.semilogy(GkT_list, trans_rate_G(GkT_list, l_list[x]), linewidth=2, color=colors[x], label ='$\lambda/k_{B}T = $'+str(l_list[x]))

plt.axvline(x=0, color='black', linewidth=1)
#plt.axvline(x=2, color='grey', linewidth=1, linestyle='--')
#plt.axvline(x=-2, color='grey', linewidth=1, linestyle='--')
#plt.axvline(x=-5, color='grey', linewidth=1, linestyle='--')

plt.legend(fontsize=10, loc = 'upper right')

plt.xlabel('$\Delta G_{0}/k_{B}T$', fontsize=10)
plt.ylabel('$k_{ij} \cdot h \cdot k_{B}T / |I_{ij}|^2 \cdot 2\pi^{3/2}$', fontsize=10)

plt.tick_params(labelsize=10, direction='in', axis='both', which='major', length=8, width=2)
plt.tick_params(labelsize=10, direction='in', axis='both', which='minor', length=4, width=2)

plt.ylim(0.0001, 4)
plt.xlim(-16, 11);
```

### Activation Energy

```python
def Delta_G_G(G, l):
    # G = list of values
    # l = number
    
    delta_G = []
    
    for g in G:
        del_G = (g + l)**2 / (4*l)
        delta_G.append(del_G)
        
    return delta_G
```

```python
Delta_G_list = np.arange(-20, 10, 0.1)
G_list = [0.1, 0.5, 2, 6, 10]
colors = ['#ffdd00', '#ff851e', '#ff0057', '#ba0087', '#0015a1']
```

```python
plt.figure(figsize=(8,6), dpi=300)

for x in range(len(G_list)):
    plt.plot(Delta_G_list, Delta_G_G(Delta_G_list, G_list[x]), linewidth=2, color=colors[x], label='$\lambda = $'+str(G_list[x]))

plt.legend(fontsize=15)

plt.xlabel('$\Delta G_{0}$', fontsize=14)
plt.ylabel('$\Delta G$', fontsize=14)

plt.tick_params(labelsize=15, direction='in', axis='both', which='major', length=8, width=2)
plt.tick_params(labelsize=15, direction='in', axis='both', which='minor', length=4, width=2)

plt.ylim(0, 10)
plt.xlim(-18, 8);
```
